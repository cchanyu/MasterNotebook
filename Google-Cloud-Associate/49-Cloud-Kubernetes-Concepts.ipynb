### Cloud GKE and Kubernetes Concepts<br>

***

<br>

Google's Kubernetes Engine (GKE) Platform as a Service(PaaS)<br>

**What is Kubernetes?**<br>

Kubernetes is an orchestration platform for containers<br>
invented by Google and open source. Now maintained by CNCF (Cloud Native Computing Foundation).<br>
automate / schedule / run containers on clusters of physical or vitual machines<br>
thus eliminating many of manual processing involved in scaling containerized applications<br>
kubernetes manage the containers that runs the application and ensure no downtime<br>

kubernetes provide you with the framework that work resiliently.<br>
it takes scaling and fail over for your application, provides the deployment patterns<br>
allows you to manage your app with flexibility, relibility and power<br>

it works with range of tools, including docker<br>

Kubernetes had its own challenges, scaling at CD, load balancing, availability,<br>
auto scaling, networking, rollback on faulty deployments<br>
<br>

**What is GKE?**<br>

Google have since developed managed offering for kubernetes, providing a managed environment<br>
for deploying, managing, and scaling your containerized applications using Google's infrastructure<br>

GKE environment consist of compute engine instances grouped together to form a cluster<br>
and it provides all the same benefits, as on-premises kubernetes, yet has abstracted the complexity<br>
having to worry about the hardware<br>

It has advanced clusters management features:<br>
- Cloud Load Balancing - being able to spread traffics amongst clusters and nodes<br>
- Node pools - to designated subnet of nodes within a cluster for additional flexibility<br>
- Automatic scaling - of your cluster node instance count<br>
- Automatic upgrades - for your cluster's node software<br>
- Node auto-repair - node health and availability<br>
- Logging and Monitoring - with Gsuite for visibility into your clusters<br>
<br>

**Cluster Architecture**<br>

A cluster is the foundation of GKE and Kubernetes as a whole.<br>
Kubernetes objects that represents your containerized applications all run on top of cluster<br>
in GKE, cluster consists of at least 1 controlled plane and multiple worker machines called nodes<br>
- one or more control planes<br>
- one or more nodes<br>
- control planes and node machines run the kubernetes clusters<br>
- **control plane** is responsible for coordinate the entire cluster<br>
this includes: scheduling workloads, like containerize applications<br> 
managing the workload lifecycle: scaling and upgrades<br>
it also manages the network and storage resources for those workloads<br>
most importantly, it manages the state of the cluster and make sure it's at the desired state<br>
- **Nodes** are the worker machines that run containerized applications and other workloads,<br>
nodes are compute engine VM instances that GKE creates on your behave when you create a cluster<br>
each node is controlled from control plane, nodes self report the status<br>
- Nodes also run the services necessary to support the docker container that make up your cluster's workloads<br>
these includes: the docker runtime, and kubernetes node agent known as **kubelet** which communicates with control plane<br>
and responsible for starting and running docker containers scheduled on that node<br>
<br>



##### **Control Plane:**<br>

Control Plane (API Server -> Kube scheduler -> Kube controller manager -> Cloud controller manager -> etcd)<br>

**API Server**<br>
- control plane is a unified **endpoint of the cluster**<br>

control planes components make global decisions about the cluster<br>
scheduling, as well as detecting and responding to cluster events<br>

- all interactions with clusters are done via kubernetes APIs call<br>

direct call: HTTP or GRPC, indirect call: **kubectl** (kubernetes commandline client)<br>
or interact with UI in Cloud console<br>
API server process is the hub for all communications in the cluster<br>

**Kube Scheduler**<br>
- components that discovers and assigns newly created pods to a node for them to run on<br>

**Kube controller manager**<br>
- components that runs controller processes<br>

also responsible for noticing and responding when nodes go down<br>
maintaining correct number of pods, populating the services in pods<br>
as well as creating default accounts, and API access tokens for new namespaces<br>

these controllers make changes to the cluster if it's not meet the desired state<br>

**Cloud controller manager**<br>
- runs controllers specific to the cloud provider<br>

this embeds cloud control specific logic. cloud controller manager let you link your cluster<br>
into any cloud providers API and separates out the components that interacts with cloud platform<br>
from the components that just interacts with clusters<br>

cloud controller manager only runs controllers that are specific to your cloud provider, in this case: Google Cloud<br>

**etcd**<br>
- key-value store that stores the state of the cluster<br>

This component is responsible for store the state of the cluster<br>
etcd is a consistent, high available key-value store that only interacts with the API server<br>

it saves all the config data along with what nodes are part of the cluster<br>
and what pods they're running<br>
<br>

##### **Node:**<br>

Node (kubelet -> kube-proxy -> runtime)<br>

**kubelet**<br>
- kubelet is an agent that runs on each node in the cluster<br>
that communicates with the control plane<br>

It's responsible for starting and running docker containers scheduled on that node<br>
it takes set of pod specs that are provided to it and ensures that it's healthy<br>

**kube-proxy**<br>
- maintains the network connectivity<br>

**runtime**<br>
- container runtime is a software that's responsible for running containers<br>
<br>

##### **GKE Abstraction**<br>

GKE manages all the control plane component<br>
endpoint exposes the kubernetes API server that kubectl uses to communicate with cluster control plane<br>
endpoint ip is displayed in cloud console and this ip will allow you to interact with the cluster<br>
> gcloud container clusters get-credentials<br>

command gets the clusters endpoint as part of updating kube config<br>
an IP for cluster then exposed to interact with, and is responsible for provisioning and managing all the infrastructure<br>
GKE also automates kubernetes nodes by launching them as Compute Engine VMs under the hood.<br>
but still allows user to change the machine type and access upgrade options<br>
