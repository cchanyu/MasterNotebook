### Cloud GKE Cluster and Node Management<br>

***

<br>

**Node Pools**<br>

Node pool is a group of nodes within a cluster with the same configuration<br>
you define the size at the creation<br>

**custom node pools** - useful for pods that require more resources<br>
you can create/upgrade/delete node pools individually without affecting the whole cluster<br>

node pools always run in latest kubernetes. Existing node pools can be manually/automatically upgraded<br>
<br>

**Cluster Types:**<br>

zonal clusters - supports both single-zone, multi-zonal<br>

**single zone cluster(zonal)** - has a single control plane running in 1 zone<br>
manages the workloads controlling nodes in 1 zone<br>

**multi zonal cluster(zonal)** - has a single replica of control plane<br>
but has nodes running in multiple zones<br>
during the outage of a zone, the cluster still runs<br>
cluster can't be configured until outage is resolved though<br>

**regional cluster(regional)** - high availability<br>
has multiple replicas of control plane, and nodes across zones<br>
it consumes more compute engine resources<br>

**private cluster** - gives you an ability to isolate nodes<br> 
from having an inbound/outbound connectivity to the public internet<br>
control plane is connected to nodes via vpc peering through internal IP addresses<br>
control plane has a private endpoint in addition to having a public endpoint<br>
<br>

**Cluster Version**<br>

Release Channel - automatic update and management of GKE<br>
Rapid, Regular(default), Stable<br>
<br>

**Rapid** - gets the latest kubernetes as fast as possible and use new GKE features<br>

**Regular(default)** - you have access to GKE and kubernetes features, soon after they're released,<br> 
2-3 months after the rapid release<br>

**Stable** - stability is prioritized over new functionality<br>
2-3 months after regular release<br>

**Specific Version** - use a specific supported ver. of kubernetes for given workloads<br>
<br>

**Cluster upgrades**<br>

Control planes and nodes DO NOT always run the same version<br>
Control planes always upgrades before its nodes<br>
- Zonal clusters - cannot launch or edit workloads during upgrade<br>
- Regional - Each control plane is upgraded one by one<br>

Auto-upgrade is enabled by default<br>
Manual upgrade - cannot upgrade control plane more than one minor version at a time<br>
Maintenance window and exclusions are available<br>
<br>

**Node and Node Pool upgrades**<br>

Auto-upgrade is enabled by default<br>
Manual upgrade is available<br>
Maintenance window and exclusions are available<br>

Pods scheduled to run on another nodes during upgrade<br>
Node is recreated in the new version, using same name<br>

Upgrade is complete only when<br>
- all nodes have been recreated<br>
- cluster is in the desired state<br>
<br>

**Surge Upgrades**<br>

Control the number of nodes GKE can upgrade at a time<br>
You can change how many nodes GKE attempts to upgrade at once by changing the surge upgrade parameter<br>
surge upgrades reduce distruption to your workload during cluster maintenance<br>
also allow you to control the number of nodes upgraded in parallel<br>
surge upgrade also work with the cluster auto scaler to prevent changes to the nodes that're being upgraded<br>

**max-surge-upgrade** - Num of additional nodes added to the node pool during an upgrade<br>
raises the number of nodes that can be upgraded simultaneously<br>
Higher number = More parallel upgrades, more money<br>

**max-unavailable-upgrade** - Num of nodes that can be simultaneously unavailable during an upgrade<br>
Higher number = More disruptive, more risk<br>

During upgrades, GKE brings down at most the sum of (max-surge-upgrade + max-unavailable-upgrade)<br>
