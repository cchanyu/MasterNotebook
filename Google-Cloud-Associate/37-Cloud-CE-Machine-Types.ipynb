### Cloud Compute Engine Machine Types<br>

***

<br>

**Machine Types**<br>

Ex: e2-standard-32<br>

e2 - series + generation<br>
series - e2,n2,n2d,n1,m1,m2,c2,f1,g1<br>

standard - type<br>
type - standard, highmem, highcpu, ultramem, megamem, medium, small, micro<br>

32 - vCPUs<br>
vCPU - 1,2,4,8,16,30,32,40,48,60,64,80,96,128,160,208,224,416<br>

GPU (ONLY available in N1)<br>
<br>

##### **pre-defined machine type families**<br>

**General Purpose**<br>

general purpose are available in standard, high memory, high cpu<br>

- standard - balance of CPU and memory<br>
- high memory - high memory to CPU ratio<br>
- high CPU - high cpu to memory ratio<br>
<br>

- **e2** - day-to-day computing at a low cost<br>
Ex: web serving, app serving, back office apps, small/medium database, microservices, virtual desktop, development environment<br>

e2 supports upto 32 vCPUs, 128 GB of memory<br>
lowest on-demand pricing<br>

e2 - standard/highmem/highcpu - 2-32 GBs<br>
min: 2 vCPU, 0.5 GB memory<br>
max: 32 vCPU, 128 GB memory<br>
<br>

- **n1/n2/n2d** - balanced price/performance across a wide range of VM shapes<br>
Ex: web serving, app serving, back office apps, medium/large databases, cache, media/streaming<br>

**n1** supports upto 96 vCPUs, 624 GB of memory<br>
only machine type for GPU and TPU support<br>
larger sustained use discount than N2<br>

n1 + gpu/tpu - standard/highmem/highcpu - 2-96 GBs<br>
min: 2 vCPU, 0.95 GB memory<br>
max: 96 vCPU, 624 GB memory<br>
<br>

**n2** supports upto 80 vCPUs, 640 GB of memory<br>
Workloads that can take advantages of the higher clock frequency<br>
Higher per thread performance<br>

n2 - standard/highmem/highcpu - 2-80 GBs<br>
min: 2 vCPU, 0.5 GB memory<br>
max: 80 vCPU, 640 GB memory<br>
<br>

**n2d** supports upto 224 vCPUs, 896 GB of memory<br>
Largest general purpose machine type<br>
Higher memory-to-core ratios<br>

n2d - standard/highmem/highcpu - 2-224 GBs<br>
min: 2 vCPU, 0.5 GB memory<br>
max: 224 vCPU, 896 GB memory<br>
<br>

**Compute Optimized**<br>

Ultra high performance for compute intensive workloads<br>

High Performance Computing (HPC)<br>
Electronic Design Automation (EDA)<br>
Gaming<br>
Single-threaded applications<br>
<br>

**c2** - standard - 4-60 GBs<br>
min: 4 vCPU, 16 GB memory<br>
max: 60 vCPU, 240 GB memory<br>

Compute intensive workloads<br>
Highest performance per core<br>
Cannot use regional persistent disks, aka only available on select zones and platforms<br>
<br>

**Memory Optimized**<br>

Ultra high-memory workloads<br>
large in-memory databases like SAP HANA<br>
In-memory analytics<br>

**m1** - ultramem/megameme - 40-96 GBs<br>
min: 40 vCPU, 32 GB memory<br>
max: 160 vCPU, 3844 GB memory<br>

**m2** - ultramem/megameme - 208-416 GBs<br>
min: 40 vCPU, 32 GB memory<br>
max: 160 vCPU, 11776 GB memory<br>

Intensive memory use<br>
In-memory databases/in-memory analytics<br>
Cannot use regional persistent disks<br>
<br>

**Shared core**<br>

E2 - micro/small/medium<br>
N1 - f1-micro/g1-small<br>

**e2** - micro/small/medium - 2 GBs<br>
min: 2 vCPU, 1 GB memory<br>
max: 2 vCPU, 4 GB memory<br>

**n1** - f1-micro/g1-small - 1 GBs<br>
min: 2 vCPU, 1 GB memory<br>
max: 2 vCPU, 4 GB memory<br>

Burstable workloads, cost-effective, non-resource intensive applications<br>
Physical core available for short periods of time<br>
CPU bursting capabilities<br>
<br>

**Custom Types**<br>

customer defined CPU and memory for custom workloads<br>
these are available for any general-purpose machine<br>

e2/n2/n2d/n1+gpu - standard/highmem/highcpu - 2-224 GBs<br>
gpu - NVIDIA Tesla K80/P4/T4/V100/P100<br>
<br>
