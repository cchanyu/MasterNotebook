### Cloud DNS Record Types<br>

***

<br>

Commonly used DNS records are saved in zone files<br>

**Name Server (NS)**<br>

root domain -> TLD name servers -> Authoritative name servers<br>
- root domain<br>
    - root zone: records for .co<br>
- TLD name servers<br>
    - .co zone<br>
        - name.co NS ns-cloud-d1.googledomains.com<br>
        - name.co NS ns-cloud-d2.googledomains.com<br>
- Authoritative name servers<br>
    - name.co zone<br>
<br>

**A and AAAA records**<br>

in name.co zone<br>
www A 52.54.92.195 (IPv4)<br>
www AAAA 2001:4860:4802:32::a (IPv6)<br>
<br>

**(Canonical Name)CNAME records**<br>

in name.co zone<br>
name.co A 52.54.92.195<br>

name.co <- shop CNAME name.co<br>
name.co <- ftp CNAME name.co<br>
both points to name.co, which points to the IP address<br>

if IP ever changes you only have to change 1 record in name.co<br>
CNAME should point to another domain name, never directly to an IP address<br>
<br>

**(Text)TXT record**<br>

type of resource that provide text information to sources outside of domain<br>
record values are either human or machine readable text<br>

many records, text records are used to verify the ownership<br>
even to provide human readable information about a server, network or data center<br>

in name.co zone<br>
TXT loves-cats<br>
TXT cats-approved<br>
<br>

**(Mail Exchange)MX records**<br>

resource record that directs email to a mail server<br>
MX server indicates how email should be routed<br>
and to which server the mail should go to<br>

in name.co zone<br>
name.co A 52.54.92.195<br>
MX 5 Mail <- 5 (priority), Mail (domain name)<br>
MX 10 ASPMX.L.GOOGLE.COM <- this is outside of google domain, a mail server<br>

laura requests the server<br>
toni@name.co -> MX query -> name.co zone<br>
uses SMTP protocol on 52.54.92.195 to deliver the mail<br>
<br>

**(Pointer)PTR records**<br>

provides domain name associated with IP address<br>
DNS pointer record is exactly the opposite of A record, which provides the IP associated with domain<br>
DNS PTR records are used to reverse dns lookup<br>

in name.co zone<br>
www A 52.54.92.195<br>
192.92.54.52.in-addr.arpa PTR name.co <- PTR record<br>

IPv6 records are also reversed and stored in arpa in different ways<br>
<br>

**(Start of Authority)SOA records**<br>

this resource will be created for you when you create manage zone<br>
and specifies the authortative informations, including global parameter of DNS zone<br>

there's 1 SOA record per zone and it includes important information about the zone<br>
